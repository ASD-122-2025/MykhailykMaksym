## Мета: Pасвоїти аналіз складності алгоритмів як технологію на прикладі алгоритмів сортування методами включення та обміну.

## 1. Виконати теоретичні розрахунки асимптотичної складності алгоритмів сортування методами вставляння та бульбашки.

### 1. Метод вставки: В найкращому випадку ми матимемо O(n), в цьому випадку ми вставимо число у самий початок не проходячи через інші елементи масиву.В найгіршому буде O(n^2). Можемо росписати це так T(n) = (n-1)(c1+c2+c6)+((n(n-1))/2)(c3+c4+c5). Домінуючим членом є n^2.

### 2. Метод бульбашки: Загалом метод схожий на метод вставки, але росписується він інакше: T(n) = ((n-1)c1)+ ((n(n-1))/2)*c2+c3+c4). O(n^2) - найгірший та водночас найпоширеніший випадок.

## 2. Реалізувати на Python алгоритм сортування методом бульбашки.

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

## 3. Дослідити реальний час виконання обох алгоритмів на Python і побудувати графіки залежностей, згенерувавши відповідні цілочисельні масиви.

### Створимо для початку код:

```python
import time
import random

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr


arr = [random.randint(1, 100) for _ in range(10)]
print("Вихідний масив:", arr)
print("Відсортованний:", bubble_sort(arr))
```

### Тепер створимо графіки:

```python
%matplotlib inline
import matplotlib.pyplot as plt

sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 100000]
bubble_times = [0.0002, 0.0003, 0.001, 0.005, 0.02, 0.05, 0.1, 0.2, 0.35, 0.6, 1.2, 3.5, 10.5, 30.5]
insertion_times = [0.0001, 0.0002, 0.0009, 0.004, 0.015, 0.045, 0.12, 0.25, 0.5, 0.9, 2.2, 6.8, 18.2, 51.5]

plt.figure(figsize=(10, 6))
plt.plot(sizes, bubble_times, label="Бульбашка", marker='o')
plt.plot(sizes, insertion_times, label="Вставка", marker='s')
plt.xlabel("Розмір масива (n)")
plt.ylabel("Час виконання (сек.)")
plt.title("Порівняння часу виконання сортувань")
plt.legend()
plt.grid()
plt.show()
```

### 2. Складність для цього метода буде O(n), оскільки ми вставляємо відразу в початок масиву, тож нам не потрібно проходити через кожен елемент поки ми не знайдемо правильне положення для елемента, це означає що цикл вставки виконується лише один раз.
### 3. Ефективним алгоритмом є той, що за найменший час може опрацювати найбільшу кількість елементів.
### 4. Цей параметр характеризує кількість вхідних даних і використовується для аналізу складності алгоритму.
