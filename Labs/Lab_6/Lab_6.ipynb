## 2. Написати функцію pop_n(), що видаляє елементи стека з його початку до номера n включно;

```python
def pop_n(stack, n):
    for i in range(n):
        if stack:
            stack.pop()
        else:
            break
```

### Insert - Асимптотична складність цього алгоритму дорівнює О(1), алгоритм вставки завжди вставляє новий елемент до вершини стеку, отже час мінімальний
### Delete - Асимптотична складність цього алогритму дорівнює О(1) по тій же причині, що і алгоритм вставки, алгоритм видалення завжди працює з верхнім елементом стеку.

## 4. Написати функцію print_n(), що друкує елементи черги з його початку до номера n включно;

```python
def print_n(queue, n):
     for i in range(min(n + 1, len(queue))):
        print(queue[i])
```

## 6. Контрольні запитання

### 1. Стек - структура даних, з якою ми можемо виконувати дії: додавання елементів, видалення елементів, пошук, шукати розімр стеку, інше.
### 2. Стек і черга відрізняються принципами роботи, оскільки стек працює за правилом "останній прийшов - перший пішов", а черга - "перший прийшов - перший пішов" 
### 3. Реалізація через масив має такі переваги як: простота, краще використання пам'яті, швидший доступ до елементів, інше, але є серйозний недолік у вигляді фіксованого розміру стеку, оскільки, якщо змінювати розмір стеку то пам'ять буде неефективно використовуватися. зв'язний список же, навпроти, має динамічний розмір, але є складнішим, отже використовує більше пам'яті, а також трохи повільніший.
### 4. Стек у програмуванні - виклик функцій, пошук у глибину (DFS), Аналіз виразу. Стек у реальному житті - стопка тарілок або чогось іншого, історія пошуку у браузері. Черга у програмуванні - планування завдань, обробка запитів, ширина пошуку (BFS). Черга у житті - черга у магазині, обслуговування клієнтів у ресторані.
